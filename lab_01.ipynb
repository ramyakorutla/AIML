matplotlib

```python
people=['akshay','ramya','chandana','radhika','bhavya','nazeera']
age=[25,19,35,45,57,69]
weight=[50,42,34,55,47,38]
height=[175,160,165,132,155,145]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(age, weight,height)
plt.show()
```

Bar chart

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.4,color='lavender',edgecolor='k',alpha=0.9)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=16,rotation=30)
plt.yticks(fontsize=16)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))

#main plot function 'hist'
plt.hist(weight,color='yellow',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

#main plot function 'hist'
plt.hist(height,color='yellow',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient height",fontsize=18)
plt.xlabel("height in cms",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
